This document specifies a content-based identifier for mappings.

# Introduction

Mappings (also known as *terminology mappings*, *correspondences*, or *semantic alignments*) are relevant to integrate and to compare data from diverse sources. If mappings are managed in distributed environments, there is a need to reference individual mappings with an identifier. Two kinds of mapping identifiers exist:

-   **Assigned identifiers** must explicitly be generated once by the source of a mapping and carried along with it.
-   **Derived identifiers** (also known as *content-based identifiers*) can be generated automatically from the content of a mapping.

This specification defines a derived identifier for mappings. One use case is the distributed review of mappings: if some mapping is reviewed, the review should apply to other mappings with same content, independent from when, where, and by whom the mapping was created.

This specification consists of a definition of [mapping content](#mappings) and the definition of a method to [derive an identifier](#identifier) from the content. The method does not depend on any specific data format or programming language.

## Mapping

A **mapping** is a 4-tuple of:

-   **subjects**: a non-empty set of IRI
-   **objects**: a non-empty set of IRI
-   **predicate**: an absolute IRI, also known as mapping type
-   **truth value**: either true for affirmative mappings or false for negative mappings

IRIs must be absolute as specificied in [RFC 3987](https://tools.ietf.org/html/rfc3987). IRIs SHOULD be based on NFC-normalized Unicode strings (see [RFC 3987](https://tools.ietf.org/html/rfc3987) section 5.3.2.4).

A mapping is also called **1-to-1 mapping**, if both subjects and objects contain exactely one IRI each.

> **Informative Notes**
>
> Applications may restrict or extends mappings, for instance by:
>
> -   limitation to 1-to-1-mappings
> -   limitation of the set of allowed predicates
> -   limitation to affirmative mappings
> -   allowing subjects and/or objects to be empty sets
>
> Handling these models of mappings is out of the scope of this specification. The same applies to resolving different IRIs to one form if they refer to the same concept.

> **Example 1: An affirmative 1-to-1 mapping**
>
> -   subjects: `http://example.org/feline`
> -   objects: `http://example.com/cat`
> -   predicate: `http://www.w3.org/2002/07/owl#sameAs`
> -   truth value: `true`

> **Example 2: A more complex mapping**
>
> -   subjects: `http://example.org/red` and `http://example.org/blue`
> -   objects: `http://example.com/green`
> -   predicate: `http://www.w3.org/2004/02/skos/core#closeMatch`
> -   truth value: `false`

# Identifier

A **mapping sameness identifier** is created from a Unicode string consisting of three elements, separated by space (U+0020):

1.  mapping subjects sorted by Unicode Code Point and separated by vertical bar (U+007C)
2.  mapping predicate
3.  mapping objects sorted by Unicode Code Point and separated by vertical bar (U+007C)

The resulting string is encoded in UTF-8, then hashed with SHA-256 ([RFC 6234](https://tools.ietf.org/html/rfc6234)), and the digest is serialized in base64url encoding without padding `=`. A tilde (`~`, U+007E) is prepended if the mapping is negative.

To use a mapping sameness identifier as IRI, the generated string is appended to URI namespace string `mapping:` (TODO: define an official namespace)

> **Warning**
>
> Some programming languages, such as Python, sort strings by Unicode Code Points. JavaScript sorts by UTF-16 value instead. If IRIs don’t include characters in the range U+010000 to U+10FFFF there is no difference.

> **Identifier of example 1**
>
> The string is `http://example.org/feline http://www.w3.org/2002/07/owl#sameAs http://example.com/cat`.
>
> The SHA-256 digest (hex) is `95a088082ab2b2a68638aebbcc3fe3e0f229da75a8b5bdbb9f3f8cd5e1e4286e`
>
> The mapping sameness identifier is `mapping:laCICCqysqaGOK67zD_j4PIp2nWotb27nz-M1eHkKG4`.

> **Identifier of example 2**
>
> The string is `http://example.org/blue|http://example.org/red http://www.w3.org/2004/02/skos/core#closeMatch http://example.com/green`.
>
> The SHA-256 digest (hex) is `424e7a86ea29d5a0aaf1d3d7da9a864b48121ac465c67163aef56f6f87bb1ba8`
>
> The mapping sameness identifier is `mapping:~Qk56huop1aCq8dPX2pqGS0gSGsRlxnFjrvVvb4e7G6g`

## JSKOS example

Mappings can be expressed in JSKOS data format with its [Mapping Object](https://gbv.github.io/jskos/#mapping):

``` json
{
  "type": [
    "http://www.w3.org/2004/02/skos/core#mappingRelation",
    "http://www.w3.org/2002/07/owl#sameAs"
  ],
  "from": {
    "memberSet": [ { "uri": "http://example.org/feline" } ]
  },
  "to": {
    "memberSet": [ { "uri": "http://example.com/cat" } ]
  }
}
```

Note that in this example <http://www.w3.org/2004/02/skos/core#mappingRelation> does not encode the mapping predicate but the fact that this JSON object is a JSKOS mapping!

JSKOS supports mapping sources and mapping objects with multiple source but negative mappings are not supported yet (see [this issue](https://github.com/gbv/jskos/issues/173)). A negative mapping will likely be expressed like this:

``` json
{
  "type": [
    false,
    "http://www.w3.org/2004/02/skos/core#closeMatch"
  ],
  "from": {
    "memberSet": [ 
      { "uri": "http://example.org/red" },
      { "uri": "http://example.org/blue" }
    ]
  },
  "to": {
    "memberSet": [ { "uri": "http://example.org/green" } ]
  }
}
```

## RDF examples

A non-negative 1-to-1 mapping can be expressed in form of an RDF triple with mapping subjects being the RDF triple subject, mapping objects being the RDF triple object, and mapping predicate being the RDF triple predicate. An RDF triples can only represent a mappings if all of its elements are absolute IRIs.

Example: the RDF statement

``` ttl
<http://example.org/A> <http://www.w3.org/2002/07/owl#sameAs> <http://example.com/B> .
```

Has mapping sameness identifier …(TODO)…

Expressing negative mappings and mappings beyond 1-to-1 mappings in RDF require more complex features and conventions such as reification or triple terms (also known as RDF-star, introduced in RDF 1.2). The choice of these features to express mappings is out of the scope of this document. The combination of mapping sameness identifiers and RDF statements to *express* mappings beyond non-negative 1-to-1 mappings is an anti-pattern:

``` ttl
# Don't do this, it's a contradiction:
<some:mapping:sameness:id:of:non:negative:mapping> a ontology:NegativeMapping .
```

## Implementations

### Python

``` python

import hashlib, base64
import sys, json

def mapping_sameness_identifier(mapping):
    subjects = "|".join(sorted(mapping["subjects"]))
    objects = "|".join(sorted(mapping["objects"]))
    string = f"{subjects} {mapping['predicate']} {objects}"

    print(string)
    sha = hashlib.sha256(string.encode('utf-8'))
    print(sha.hexdigest())
    id = base64.urlsafe_b64encode(sha.digest()).decode('ascii').rstrip('=')
    return f"mapping:~{id}" if mapping["negative"] else f"mapping:{id}"

if __name__ == "__main__":
    mapping = json.load(sys.stdin)
    print(mapping_sameness_identifier(mapping))

```

### Command line tools

``` sh

#!/bin/bash
export LANG=C.UTF-8

read -r s p o n <<<$(jq -r '[
    (.subjects|sort|join("|")),
    .predicate,
    (.objects|sort|join("|")),
    (if .negative then "~" else "" end)
  ]|join(" ")')

digest=$(echo -n "$s $p $o" | sha256sum | cut -f1 -d' ') 
base64=$(echo $digest | xxd -r -p | basenc --base64url | tr -d =)

echo "mapping:$n$base64"

```
